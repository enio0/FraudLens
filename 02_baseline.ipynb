## Why accuracy is the wrong metric here

The dataset is 99.83% legit, 0.17% fraud. A "model" that always
predicts *not fraud* would be 99.83% accurate while catching zero
fraud — useless for the actual task. Before building anything, we
prove this to ourselves, then switch to precision / recall / AUPRC,
which actually pay attention to the rare class.

```python
import pandas as pd

df = pd.read_csv("data/raw/creditcard.csv")
df.shape
```

## The lazy baseline: "always predict not fraud"

Accuracy = (correct predictions) / (total). If we always guess *not
fraud*, every legit row is "correct" and every fraud row is "wrong."
So accuracy = (# legit) / (# total), no model required. We compute
it straight from the labels to see how high a useless model can score.

```python
n_total = len(df)
n_legit = (df["Class"] == 0).sum()

lazy_accuracy = n_legit / n_total
print(f"Legit rows: {n_legit}")
print(f"Total rows: {n_total}")
print(f"Lazy 'always not-fraud' accuracy: {lazy_accuracy:.4%}")
```

## Train/test split

We evaluate the model on data it never trained on, so we hold back a
20% test set. Because fraud is only 0.17%, a naive random split could
deal fraud unevenly between the halves. We **stratify on `Class`** so
both train and test preserve the same fraud ratio, the test set stays
a fair miniature of reality.

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["Class"])   # features: everything except the label
y = df["Class"]                  # label: fraud (1) or legit (0)

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% held back for testing
    stratify=y,           # preserve the fraud ratio in both halves
    random_state=42,      # reproducible split
)

print("Train fraud rate:", y_train.mean())
print("Test fraud rate: ", y_test.mean())
```
